### Función para quitar las tildes y normalizar caracteres de la Columna MUNICIPIO

```python
#Dentro de nuestro problema de investigación conocer la cobertura por Departamento/Municipio son variables importantes por lo que debemos garantizar su correcto Nombre.
# Función para quitar las tildes y normalizar caracteres de la Columna MUNICIPIO
def remove_accents(text):
    nfkd = unicodedata.normalize('NFKD', text)
    return ''.join([c for c in nfkd if not unicodedata.combining(c)])

# Procesar la columna "MUNICIPIO"
df['MUNICIPIO'] = df['MUNICIPIO'].str.lower()  # Convertir a minúsculas
df['MUNICIPIO'] = df['MUNICIPIO'].str.replace(r'\s+', ' ', regex=True)  # Eliminar espacios extra
df['MUNICIPIO'] = df['MUNICIPIO'].apply(remove_accents)  # Eliminar tildes
#_______________________________________________________________________________________________________________________________________________________________________________________________________________________________

# Función para quitar las tildes y normalizar caracteres de la Columna DEPARTAMENTO
def remove_accents(text):
    nfkd = unicodedata.normalize('NFKD', text)
    return ''.join([c for c in nfkd if not unicodedata.combining(c)])

# Procesar la columna "DEPARTAMENTO"
df['DEPARTAMENTO'] = df['DEPARTAMENTO'].str.lower()  # Convertir a minúsculas
df['DEPARTAMENTO'] = df['DEPARTAMENTO'].str.replace(r'\s+', ' ', regex=True)  # Eliminar espacios extra
df['DEPARTAMENTO'] = df['DEPARTAMENTO'].apply(remove_accents)  # Eliminar tildes
```

```python
## Convertimos a enteros (Eliminando el punto y el cero al final) de la columnas para evitar errores en la concatenación
df['COD DEPARTAMENTO'] = df['COD DEPARTAMENTO'].astype(int)
df['COD MUNICIPIO'] = df['COD MUNICIPIO'].astype(int)
```

### Concatenar las columnas "COD DEPARTAMENTO" y "COD MUNICIPIO" con un guion bajo

```python
# Concatenar las columnas "COD DEPARTAMENTO" y "COD MUNICIPIO" con un guion bajo
df['Llave'] = df['COD DEPARTAMENTO'].astype(str).str.zfill(4) + '_' + df['COD MUNICIPIO'].astype(str).str.zfill(4)
```

### Dentro de nuestro problema de investigación conocer la cobertura por Departamento/Municipio son variables importantes por lo que debemos garantizar su correcto Nombre.
### Función para quitar las tildes y normalizar caracteres de la Columna MUN_REV_IN y DEP_REV_IN

```python
#Dentro de nuestro problema de investigación conocer la cobertura por Departamento/Municipio son variables importantes por lo que debemos garantizar su correcto Nombre.
# Función para quitar las tildes y normalizar caracteres de la Columna MUN_REV_IN y DEP_REV_IN
def remove_accents(text):
    nfkd = unicodedata.normalize('NFKD', text)
    return ''.join([c for c in nfkd if not unicodedata.combining(c)])

# Procesar la columna "MUN_REV_IN"
df_revistas['MUN_REV_IN'] = df_revistas['MUN_REV_IN'].str.lower()  # Convertir a minúsculas
df_revistas['MUN_REV_IN'] = df_revistas['MUN_REV_IN'].str.replace(r'\s+', ' ', regex=True)  # Eliminar espacios extra
df_revistas['MUN_REV_IN'] = df_revistas['MUN_REV_IN'].apply(remove_accents)  # Eliminar tildes

# Procesar la columna "DEP_REV_IN"
df_revistas['DEP_REV_IN'] = df_revistas['DEP_REV_IN'].str.lower()  # Convertir a minúsculas
df_revistas['DEP_REV_IN'] = df_revistas['DEP_REV_IN'].str.replace(r'\s+', ' ', regex=True)  # Eliminar espacios extra
df_revistas['DEP_REV_IN'] = df_revistas['DEP_REV_IN'].str.replace(',', ' ')  # Eliminar comas
df_revistas['DEP_REV_IN'] = df_revistas['DEP_REV_IN'].apply(remove_accents)  # Eliminar tildes
```

```python
display(df_revistas)
```

Create mappings for municipio and departamento to their codes

```python
# Create mappings for municipio and departamento to their codes
municipio_mapping = df.set_index('MUNICIPIO')['COD MUNICIPIO'].to_dict()
departamento_mapping = df.set_index('DEPARTAMENTO')['COD DEPARTAMENTO'].to_dict()

# Function to apply the mappings and handle missing values
def map_codes(row):
    municipio = row['MUN_REV_IN']
    departamento = row['DEP_REV_IN']

    cod_municipio = municipio_mapping.get(municipio)
    cod_departamento = departamento_mapping.get(departamento)

    return pd.Series({'COD_MUNICIPIO': cod_municipio, 'COD_DEPARTAMENTO': cod_departamento})

# Apply the mapping function to create new columns
df_revistas[['COD_MUNICIPIO', 'COD_DEPARTAMENTO']] = df_revistas.apply(map_codes, axis=1)
```

Se agrupa  el conteo de nulos en cod_departamento por cada valor en DEP_REV_IN
en busca de valores que puedan estan dando problema

```python
"""Se agrupa  el conteo de nulos en cod_departamento por cada valor en DEP_REV_IN
en busca de valores que puedan estan dando problemas"""
null_counts = df_revistas.groupby('DEP_REV_IN')['COD_DEPARTAMENTO'].apply(lambda x: x.isnull().sum())

# Display the result
null_counts
```

Se reemplaza el valor de bogotá en el dataframe de revistas, columna de departamento por el valor de bogota en la columna de
deoartamento del dataframe de cobertura movil y se rellena el codigo de departamento por el
valor correspondient

```python
"""Se reemplaza el valor de bogotá en el dataframe de revistas, columna de departamento por el valor de bogota en la columna de
deoartamento del dataframe de cobertura movil y se rellena el codigo de departamento por el
valor correspondiente"""
df_revistas['DEP_REV_IN'] = df_revistas['DEP_REV_IN'].replace('bogota d. c.', 'bogota d.c')

# Replace null values in 'COD_DEPARTAMENTO' with 11001 when 'DEP_REV_IN' contains "bogota"
df_revistas.loc[df_revistas['DEP_REV_IN'].str.contains('bogota'), 'COD_DEPARTAMENTO'] = df_revistas.loc[df_revistas['DEP_REV_IN'].str.contains('bogota'), 'COD_DEPARTAMENTO'].fillna(11)
```

```python
# Fromateo correctamente las columnas de cod_municipio y cod_departamento
df_revistas['COD_MUNICIPIO'] = df_revistas['COD_MUNICIPIO'].astype(str).str.split('.').str[0]
df_revistas['COD_DEPARTAMENTO'] = df_revistas['COD_DEPARTAMENTO'].astype(str).str.split('.').str[0]
```

### Errores de Formato

```python
def validar_tipos(df):
    """
    Validar tipos de un df
    """
    errores = []

    for columna, info in cols_info.items():
        if columna in df.columns:
            tipo_real = df[columna].dropna().map(type).unique()[0]
            tipo_correcto = issubclass(tipo_real, tuple(tipo_esperado[info["tipo_dato"]]))

            if not tipo_correcto:
                errores.append(f"❌ La columna [{columna}] debe ser {info['tipo_dato']}, pero tiene {tipo_real}")

    if errores:
        print("\n".join(errores))
    else:
        print("✅ Todos los tipos de datos son correctos.")
```
